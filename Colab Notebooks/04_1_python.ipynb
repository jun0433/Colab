```
# List indexing

my_list = ['p', 'r', 'o', 'b', 'e']

# Output: p
print(my_list[1])

# Output: o
print(my_list[2])

# Output: e
print(my_list[4])
```

```
# Nested List
n_list = ["Happy", [2, 0, 1, 5]]

# Nested indexing
# n_list[0] 가 List (char의 sequence)
print(n_list[0][1]) # a

print(n_list[1][3]) # 5

# Error! Only integer can be used for indexing
#print(my_list[4.0])
```

```
# Negative indexing in lists
my_list = ['p','r','o','b','e']

print(my_list[-1]) # e

print(my_list[-5]) # p
```

```
# List slicing in Python

my_list = ['p','r','o','g','r','a','m','i','z']

# elements 3rd to 5th
print(my_list[2:5]) # o g r 

# elements beginning to 4th
print(my_list[:-5]) # p r o g

# elements 6th to end
print(my_list[5:]) # a m i z

# elements beginning to end
print(my_list[:])
```

```
# Deleting list items
my_list = ['p', 'r', 'o', 'b', 'l', 'e', 'm']

# delete one item
del my_list[2]

print(my_list)

# delete multiple items
del my_list[1:5]

print(my_list)

# delete entire list
del my_list

# Error: List not defined
# print(my_list)
```

```
my_list = ['p','r','o','b','l','e','m']
my_list.remove('p')

# Output: ['r', 'o', 'b', 'l', 'e', 'm']
print(my_list)

# Output: 'o'
print(my_list.pop(1))

# Output: ['r', 'b', 'l', 'e', 'm']
print(my_list)

# Output: 'm'
print(my_list.pop())

# Output: ['r', 'b', 'l', 'e']
print(my_list)

my_list.clear()

# Output: []
print(my_list)
```

```
# Python list methods
my_list = [3, 8, 1, 6, '0', 8, 4]

# Output: 1
print(my_list.index(8))

# Output: 2
print(my_list.count(8))

my_list.sort(key=int)

# Output: [0, 1, 3, 4, 6, 8, 8]
print(my_list)

my_list.reverse()

# Output: [8, 8, 6, 4, 3, 1, 0]
print(my_list)
```
