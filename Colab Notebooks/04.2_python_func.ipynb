```
total = 0
f1, f2 = 1, 2

while f1 < 4000000:
    if  f1 % 2 == 0:
      total = total + f1
    f1, f2 = f2, f1 + f2

print(total)
```

```
def hello_func(names):
  for n in names:
    print("Hello, ", n, "!")
  names[0] = 'Susie'
  names[1] = 'Pete'
  names[2] = 'Will'

org_names = ['Susan', 'Peter', 'William']
hello_func(org_names)

print("The names are now ", org_names, ".")
```
