```python
# 문제1. 한줄로 리스트 만들기 [list comprehension]
# * 리스트 컴프리헨션(list comprehensio)은 쉽게 말해 ‘리스트를 쉽게, 짧게 한 줄로 만들 수 있는 파이썬의 문법’이다.
# 1부터 1000까지의 숫자 중에 7로 나누어 떨어지는 수를 리스트 컴프리헨션을 이용하여 만드시오.

[list for list in range(1000) if list % 7 == 0]
```

```python
# 문제2. 숫자 리스트를 입력 받아서, 절대값을 기준으로 정렬(sort)한 결과를 내는 함수 absort를 작성하고, 
# 아래 예제를 실행하시오
# 실행 : absort([2, -1, 3, -4, -2, 1])
# 결과 : [-1, 1, -2, 2, 3, -4]


def absort(number): 
  number = sorted(number, key=lambda num:int(num))
  number = sorted(number, key=lambda num:abs(num))
  print(number)

absort([2, -1, 3, -4, -2, 1])

```

```python
# 문제3. 숫자 리스트를 입력 받아서, 동일한 값을 제거한 리스트를 리턴하는 함수를 작성하고
# 아래 예제를 실행하시오
# 실행 : uniq([1, 3, 2, 5, 2, 1])
# 결과 : [1, 2, 3, 5]

def uniq(number):
  print(set(number))

uniq([1, 3, 2, 5, 2, 1])
```

```python
# 문제4. 문자열을 입력받은 후 해당 입력에 각 char가 나타난 횟수를 dictionary 타입으로 리턴하는 함수를 작성하고
# 아래 예제를 실행하시오

# 실행 : make_freq('Myongji University')
# 결과 : {'M':1, 'y':2, 'o':1, 'g':1, 'j':1, 'i':3, 'U':1, 'n':1, 'v':1, 'e':1, 'r':1, 's':1, 't':1}

def make_freq(str):
  d = dict()
  str = str.replace(" ","")
  for i in str:
    if i in d:
      d[i] += 1
    else:
      d[i] = 1
  print(d)



make_freq('Myongji University')


```
