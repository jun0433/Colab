```python
from sklearn.datasets import load_iris

#load_iris 데이터를 dataset에 할당
dataset = load_iris()
```

```python
##dataset의 target shape를 확인
print(dataset.target.shape)
```

```python
#data와 target을 학습 data와 label로 선언
train_data = dataset.data
train_labels = dataset.target
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers

#신경망을 만듦
model = models.Sequential()
#입력 4, 출력 16의 'relu' activation을 사용
model.add(layers.Dense(16, activation='relu', input_shape=(4,)))
#입력 16, 출력 3의 'softmax' activation을 사용
model.add(layers.Dense(3, activation='softmax'))
```

```python
# from keras.utils import to_categorical
from tensorflow.keras.utils import to_categorical

# one-hot encoding
train_labels = to_categorical(train_labels)
```

```python
# 손실함수와 최적화, 훈련 과정을 모니터링할 지표 설정
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#신경망 훈련
model.fit(train_data, train_labels, epochs = 5, batch_size=64)
```
