# 아래 주어진 데이터셋에 대하여 딥러닝 학습을 진행하시오

우리는 mnist데이터 셋을 다소 엉뚱해 보이는 모델을 학습하고자 한다.

구현할 모델은 입력 이미지를 받아서 동일한 입력이미지를 출력하는 모델이다. 따라서, mnist에서 우리가 사용할 데이터는 x_train 이미지 데이터 뿐이다. (정답label데이터도 x_train 이미지 데이터이기 때문이다.)

1. 모델은 Dense Layer 4개
   * Dense( 128, .., input_shape=..) # relu
   * Dense( 16 ) # relu
   * Dense( 128 ) # relu
   * Dense( 28*28, activation='sigmoid') # sigmoid

2. loss는 binary_crossentropy 로 세팅,
   metrics = "mse"로 세팅

3. epochs는 20회로 세팅 

학습 완료 후에 실제도 어떻게 동작되는지 살펴보기 위해, 자신의 학번에 해당하는 이미지들을 이용하여, 입력 이미지와 출력이미지가 어떻게 차이가 나는지 가시화하시오.

```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import models


model = models.Sequential()
model.add(layers.Dense(128, activation='relu', input_shape=(28 * 28,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(28*28, activation='sigmoid'))
```

```python
model.compile(optimizer='rmsprop',
                loss='binary_crossentropy',
                metrics=['mse'])
```

```python
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255
```

```python
# from keras.utils import to_categorical
from tensorflow.keras.utils import to_categorical

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
model.fit(train_images, train_images, epochs=20, batch_size=128)
```
